```python
datatype = "MeTDiff"
m6add_df = get_m6add_files(datatype)
m6add_df
```

```python
# Normalize chromosome format
m6add_df['chr'] = m6add_df['chr'].astype(str).str.lower().str.replace('chr', '')

# Use midpoint of start-end range in GSE as approximate genome position
m6add_df['genome_pos'] = ((m6add_df['start'] + m6add_df['end']) / 2).astype(int)

m6add_df
```

```python
def fuzzy_merge(m6anet_df, m6add_df, window=3):
    matches = []
    for idx, row in m6anet_df.iterrows():
        chr_match = m6add_df['chr'] == row['chromosome']
        pos_match = (m6add_df['genome_pos'] >= row['genome_pos'] - window) & \
                    (m6add_df['genome_pos'] <= row['genome_pos'] + window)
        match_df = m6add_df[chr_match & pos_match]
        for _, match_row in match_df.iterrows():
            combined = row.to_dict()
            combined.update({f'm6add_{col}': match_row[col] for col in m6add_df.columns})
            matches.append(combined)
    return pd.DataFrame(matches)
```

```python
# Run fuzzy matching
matched_df = fuzzy_merge(m6anet, m6add_df, window=10)
```
